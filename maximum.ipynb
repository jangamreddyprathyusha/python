1.find 1st max value

```python
l=[9,8,10,4,13,1,13,10,9]
max=l[0]
for i in range(1,len(l)):
    if(max<l[i]):
        max=l[i]
print(max)
```

2.find 2nd max value

```python
l=[9,8,10,4,13,1,13,10,9]
max=0
smax=0
for i in range(len(l)):
    if(max<l[i]):
        smax=max
        max=l[i]
    elif(smax<l[i] and l[i]<max):
        smax=l[i]
print(smax)
```
