7.count number of 1's & no.of zeros 0's

```python
n=10
count1=0
count0=0
while n!=0:   #10!=0
    if n%2==1:  #10%2=>0==1,5%2=>1==1,2%2=>0==1,1%2=>1==1
        count1+=1  #0,1,1,2
    else:
        count0+=1  #1,1,2,2
    n=n//2       #10//2=5,5//2=2,2//2=1,1//2=0
print(count1)
print(count0)        
```

8.find position of 1's   x=1010--->(3,1),x=1110--->(3,2,1)

```python
n=10
pos=[]
i=0
while n!=0:               #6!=0                             #10!=0
    if n%2==1:            #6%2=>0==1,3%2=>1==1,1%2=>0==1    #10%2=>0==1,5%2=>1==1,2%2=>0==1,1%2=>1==1
        pos.append(i)     #1,2                              #1,3
    n//=2                 #6//2=3,3//2=1,1//2=0             #10//2=5,5//2=2,2//2=1,1//2=0
    i+=1                  #0+1=1,1+1=2,2+0=2                #0+1=1,1+1=2,2+0=2,2+1=3
print(pos)                
```

left and right sum difference
#left    right   diff    value       
# 0       4,8,3    0-(4+8+3)  0-15   abs(-15)    15
#10       8,3       10-(8+3)   10-11   ats(-1)    1
#10,4      3         14-3       11      ats(11)    11
#10,4,8     0         22-0       22      ats(22)    22    

```python
nums=[10,4,8,3]
sum1=sum(nums)
l=[]
left=0
for i in range(len(nums)):
    x=abs(left-(sum1-nums[i]-left))
    l.append(x)
    left+=nums[i]
print(l)

```

#using slicing left and right sum difference

```python
nums=[10,4,8,3]
l=[]        
for i in range(len(nums)):
    L=sum(nums[:i])
    R=sum(nums[i+1:])
    l.append(abs(L-R))
print(l)
```

```python
n=5
count=0
for i in range(1,n+1):
    if(n%i==0):
        count+=1
if(count==2):
    print("prime")
else:
    print("not prime")
```

```python
l=[1,2,3,4,5]
max=0
sum=0
for i in range(len(l)):
    for j in range(i+1,len(l)):
        sum=l[i]+l[j]
        count=0
        for k in range(1,sum+1):
            if(sum%k==0):
                count+=1
        if(count==2 and max<sum):
            max=sum
print(max)
        
```

multiple of the two numbers,using bitwise,even,min

```python
l=[5,3,4,1,6]
prod=1
min=1000
for i in range(len(l)): 
    for j in range(i+1,len(l)):
        prod=l[i]*l[j]
        if(prod&1==0) and (min>prod):
            min=prod
print(min)          
```

```python
nums= [2,5,1,3,4,7]
n=3
ans = []
for i in range(n): #i>0
    ans.extend([nums[i], nums[i+n]])         #num[0],num[0+3]=[2,3]
print(ans)                                   #num[1],num[1+3]=[5,4]
                                             #num[2],num[2+3]=[1,7]
```
