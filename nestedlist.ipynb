nested list

```python
l=[[10,20,[40,50]]]
for i in l:
    print(i)
```

```python
l=[[10,20,[40,50]]]
for i in l:
    for j in i:
       print(j)
```

```python
l=[[10,20,[30,["ravi","1234"]]]]
for i in l:
    print(i)
```

```python
l=[[10,20,[30,["ravi","1234"]]]]
for i in l:
    for j in i:
       print(j)
```

```python
l=[[10,20,[30,["ravi","1234"]]]]
for i in l:
    for j in i:
        if isinstance(j,list):
            for k in j:
                if isinstance(k,list):
                    for m in k:
                        print(m)
                else:
                    print(k)
        else:    
            print(j)
```
