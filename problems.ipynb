7.  find strong number 145
1.separate individual digit-1 4 5
2.find factorial    #1!=1,4!=24,5!=120
3.sum of fact      #1+24+120=145
4.condition check  

```python
n=145
m=n
sum=0
while n>0:
    temp=n%10
    fact=1
    while(temp>0):
        fact*=temp
        temp-=1
    sum+=fact
    n//=10
if(sum==m):
    print(m,"strong number")
else:
    print(m,"not strong number")
```

```python
n=153
sum=0
length=len(str(n))
r=n
while n>0:
    temp=n%10
    sum+=temp**length
    n//=10
if sum==r:
    print(r,"Armstrong number")
else:
    print(r,"Not an Armstrong number")
```

```python
#unique number of occurences like true or false
arr=[1,4,2,2,3,4,5,3]
d={}
for key in arr:
    d[key]=d.get(key,0)+1
l=list(d.values())
uniq=set(l)
print(uniq)
for i in uniq:
    if l.count(i)>1:
        print("False")
        break
else:
    print("True")
```
