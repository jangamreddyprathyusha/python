before the loop 

```python
l=[-1,2,8,-7,3,-6,2,3]
neg=[]
pos=[]
l=[neg.append(val) if val<0 else pos.append(val) for val in l]
print(neg)
print(pos)
```

after loop check the condition

```python
l=[-1,2,8,-7,3,-6,2,3]
neg=[val for val in l if val<0]
pos=[val for val in l if val>0]
print(neg)
print(pos)
```

task-->separate vowels and consonents in string "hi hello ravi garu"

```python
s = "hi hello ravi garu"
vow=[ch for ch in s if ch.lower() in "aeiou"]
con=[ch for ch in s if ch.isalpha() and ch.lower() not in "aeiou"]
print(vow)
print(con)
```
