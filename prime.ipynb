count the prime

```python
primecount=0
for n in range(1,10):
    count=0
    for i in range(1,n+1,1):
        if(n%i==0):
            count=count+1
    if(count==2):
        primecount+=1
print(primecount)
```

prime

```python
for n in range(1,10):
    count=0
    for i in range(1,n+1,1):
        if(n%i==0):
            count=count+1
    if(count==2):
        print(n,"prime")
    else:
        print(n,"not prime")
            
```

prime with even

```python
for n in range(1,10):
    count=0
    for i in range(1,n+1,1):
        if(n%i==0):
            count=count+1
    if(count==2 and n%2==0):
        print(n)
```

prime with odd

```python
for n in range(1,10):
    count=0
    for i in range(1,n+1,1):
        if(n%i==0):
            count=count+1
    if(count!=2 and n%2==1):
        print(n)
```
